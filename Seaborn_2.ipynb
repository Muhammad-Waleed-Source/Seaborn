## 4.1 Categorical Scatter Plot

### 4.1.1 Strip Plot

```python
tips.head()
```

```python
# axes level function

sns.stripplot(data=tips, x='day', y='total_bill')
```

```python
# if we want the scatter in a line
sns.stripplot(data=tips, x='day', y='total_bill', jitter=False)
```

```python
# figure level function using catplot

sns.catplot(tips, x='day', y='total_bill', kind='strip', hue='sex')
```

```python
# jitter
sns.catplot(tips, x='day', y='total_bill', kind='strip', jitter=0.3)
```

### 4.1.2 Swarm Plot

```python
# represent distribution a bit

# axes level function
sns.swarmplot(tips, x='day', y='total_bill')
```

```python
# figure level function

sns.catplot(tips, x='day', y='total_bill', kind='swarm')
```

```python
# hue
sns.catplot(tips, x='day', y='total_bill', kind='swarm', hue='sex')
```

## 4.2 Categorical Distribution Plot

### 4.2.1 Box Plot

A boxplot is a standardized way of displaying the distribution of data based on a five number summary (“minimum”, first quartile [Q1], median, third quartile [Q3] and “maximum”). It can tell you about your outliers and what their values are. Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.

image.png

```python
# axes level function

sns.boxplot(tips, x='sex', y='total_bill', hue='sex')
```

```python
sns.boxplot(tips, x='day', y='total_bill', hue='day')
```

```python
# figure level function
sns.catplot(tips, x='day', y='total_bill', hue='sex', kind='box')
```

```python
# single boxplot -> numerical col

sns.boxplot(tips, y='total_bill')

# We can see our median is somewhere around 18.
```

### 4.2.2 Violin Plot
(Boxplot + KDEplot)

```python
sns.violinplot(tips, x='day', y='total_bill', hue='day')
```

```python
sns.catplot(tips, x='day', y='total_bill', hue='day', kind='violin')
```

```python
# hue
sns.catplot(tips, x='day', y='total_bill', hue='sex', kind='violin')
```

```python
# for side by side plots

sns.catplot(tips, x='day', y='total_bill', hue='sex', kind='violin', split=True)
```

## 4.3 Categorical Estimate Plot

### 4.3.1 Bar Plot

```python
# axes level function

sns.barplot(tips, x='sex', y='total_bill', hue='sex')

# this give the avg total bill by male and female because it has a parameter called estimator with default value mean.
```

When there are multiple observations in each category, it also uses bootstrapping to compute a confidence interval around the estimate, which is plotted using error bars

```python
# To remove error bar

sns.barplot(tips, x='sex', y='total_bill', hue='sex', ci=None)
```

```python
import numpy as np

# standard deviation
sns.barplot(tips, x='sex', y='total_bill', hue='sex', estimator=np.std)

# now it is giving the standard deviation
```

```python
# min
sns.barplot(tips, x='sex', y='total_bill', hue='sex', estimator=np.min)
```

```python
sns.barplot(tips, x='sex', y='total_bill', hue='smoker', estimator=np.min)
```

### 4.3.2 point Plot

```python
sns.pointplot(tips, x='sex', y='total_bill', hue='sex', estimator=np.min)
```

```python
sns.pointplot(tips, x='sex', y='total_bill', hue='smoker', estimator=np.min)
```

```python
sns.pointplot(tips, x='sex', y='total_bill')
```

### 4.3.3 Count Plot

A special case for the bar plot is when you want to show the number of observations in each category rather than computing a statistic for a second variable. This is similar to a histogram over a categorical, rather than quantitative, variable.

The statistic here is count.

Same as histogram but histogram is on numerical data, whereas count plot is on categorical data.

```python
sns.countplot(tips, x='sex')
```

```python
sns.countplot(tips, x='sex', hue='day')
```

```python
# faciting using catplot

sns.catplot(tips, x='sex', y='total_bill', col='smoker', hue='sex', kind='box')
```

```python
sns.catplot(tips, x='sex', y='total_bill', col='smoker', row='time', hue='sex', kind='box')
```

## 5. Regression Plots

- regplot (axes level function)
- lmplot (figure level function)

In the simplest invocation, both functions draw a scatterplot of two variables, x and y, and then fit the regression model y ~ x and plot the resulting regression line and a 95% confidence interval for that regression.

### 5.1 Reg Plot

A built-in function that creates a scatter plot of two variables and fits a linear regression model through the data, helping you visualize the strength and direction of a relationship

```python
tips.head()
```

```python
# normal scatter plot

sns.scatterplot(tips, x='total_bill', y='tip')
```

```python
# scatter plot with regression line
# axes level function
sns.regplot(tips, x='total_bill', y='tip')

# the blue shade specifies the error of the line.
```

```python
# figure level function

sns.lmplot(tips, x='total_bill', y='tip')
```

```python
# hue (it cannot be applied on regplot)

sns.lmplot(tips, x='total_bill', y='tip', hue='sex')
```

## 6. Residual Plot

The residual plot is a visual map of your model's errors.Every dot on a residual plot represents one data point, showing how far off your model's guess was.

```python
sns.residplot(tips, x='total_bill', y='tip')
```
