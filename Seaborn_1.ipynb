### 1. Relational Plot

- to see the statistical relation between 2 or more variables.
- Bivariate Analysis

Plots under this section

- scatterplot
- lineplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

### 1.1 Scatter Plot

```python
# scatter plot -> axes level function

sns.scatterplot(tips, x='total_bill', y='tip')

# Here we can clearly see the x and y labels are automatically given unlike the matplotlib where we have to explicitly mention them.
```

```python
# hue (for catgorization)
# number of males and females

sns.scatterplot(tips, x='total_bill', y='tip', hue='sex')

# also a legend is created by default
```

```python
# style
# number of dinner and lunch

sns.scatterplot(tips, x='total_bill', y='tip', hue='sex', style='time')
```

```python
# size
sns.scatterplot(tips, x='total_bill', y='tip', hue='sex', style='time', size='size')
```

```python
# scatter plot -> figure level function -> square shaped -> recommended by seaborn

sns.relplot(tips, x='total_bill', y='tip', kind='scatter')
```

```python
sns.relplot(tips, x='total_bill', y='tip', kind='scatter', hue='sex')

# here we can see that figure level graphs are placing the legend outside the figure, which means they have more control.
```

```python
sns.relplot(tips, x='total_bill', y='tip', kind='scatter', hue='sex', style='time')
```

```python
sns.relplot(tips, x='total_bill', y='tip', kind='scatter', hue='sex', style='time', size='size')
```

### 1.2 Line Plot

```python
gap = px.data.gapminder()
gap.head()
```

```python
temp_df = gap[gap['country'] == 'Pakistan']
temp_df
```

```python
# axes level function

sns.lineplot(temp_df, x='year', y='lifeExp')
```

```python
# figure level function (rel plot)

sns.relplot(temp_df, x='year', y='lifeExp', kind='line')
```

```python
# hue and style

temp_df = gap[gap['country'].isin(['Pakistan', 'Brazil', 'Germany'])]

temp_df
```

```python
sns.lineplot(temp_df, x='year', y='lifeExp', hue='country')
```

```python
# countries

sns.relplot(temp_df, x='year', y='lifeExp', kind='line', hue='country')
```

```python
# continents

sns.relplot(temp_df, x='year', y='lifeExp', hue='country', style='continent', kind='line', size='continent')
```

### 1.3 Facet Plot
Facet plots in Seaborn break data into a grid of subplots (small multiples) based on categorical variables. This allows you to easily compare distributions or relationships across different subsets of data while maintaining the exact same axis scales for all panels.

Work only with figure level functions not axes level.

```python
tips.head()
```

```python
sns.relplot(tips, x='total_bill', y='tip', kind='scatter', hue='sex')
```

```python
# facet plot of above
# Col wise

sns.relplot(tips, x='total_bill', y='tip', kind='scatter', col='sex')
```

```python
# Row wise

sns.relplot(tips, x='total_bill', y='tip', kind='scatter', row='sex')
```

```python
# Both Row and Col wise

sns.relplot(tips, x='total_bill', y='tip', kind='scatter',col='smoker', row='sex')
```

```python
sns.relplot(tips, x='total_bill', y='tip', kind='line',col='sex', row='day')
```

```python
# col wrap
gap.sample(5)
```

```python
sns.relplot(gap, x='lifeExp', y='gdpPercap', kind='scatter')
```

```python
# What if we want scatter plot for every year seperatley

sns.relplot(gap, x='lifeExp', y='gdpPercap', kind='scatter', col='year')

# As we can see it is not readable, so here we use col wrap
```

```python
sns.relplot(gap, x='lifeExp', y='gdpPercap', kind='scatter', col='year', col_wrap=3)
```

### 2. Distribution Plots

- used for univariate analysis
- used to find out the distribution
- Range of the observation
- Central Tendency (mean, median etc)
- is the data bimodal? (multiple peaks)
- Are there outliers? (extremely high or low value)

Plots under distribution plot

- histplot
- kdeplot
- rugplot

**axes level:** histplot, kdeplot, rugplot

**figure level:** displot

### 2.1 Hist Plot

```python
# univariate histogram
tips.head()
```

```python
# axes level function (histplot)

sns.histplot(tips, x='total_bill')
```

```python
sns.histplot(tips, x='total_bill', bins=20)
```

```python
# figure level function (displot)

sns.displot(tips, x='total_bill', kind='hist')
```

```python
# It’s also possible to visualize the distribution of a categorical variable using the logic of a histogram.
# Discrete bins are automatically set for categorical variables

sns.displot(tips, x='day', kind='hist')
```

```python
# hue parameter

sns.displot(tips, x='tip', kind='hist', hue='sex')
```

```python
# element -> step

sns.displot(tips, x='tip', kind='hist', hue='sex', element='step')
```

```python
titanic = sns.load_dataset('titanic')
titanic.sample(5)
```

```python
# number of male and females in particular age group.

sns.displot(titanic, x='age', kind='hist', hue='sex', element='step')
```

```python
# faceting using col and row

sns.displot(titanic, x='age', kind='hist', col='sex', element='step')
```

### 2.2 KDE (Kernel Density Estimation) Plot

Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate.

Prefered compared to hist plot, because it shows all the values where as hist cannot show the in-between values.

```python
tips.head()
```

```python
# axes level function

sns.kdeplot(tips, x='total_bill')
```

```python
# figure level function

sns.displot(tips, x='total_bill', kind='kde')
```

```python
# hue

sns.displot(tips, x='total_bill', kind='kde', hue='sex')
```

```python
# hue with fill

sns.displot(tips, x='total_bill', kind='kde', hue='sex', fill=True)
```
