```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# scatter plot -> axes level function

sns.scatterplot(tips, x='total_bill', y='tip')

# Here we can clearly see the x and y labels are automatically given unlike the matplotlib where we have to explicitly mention them.
```

```python
# hue (for catgorization)
# number of males and females

sns.scatterplot(tips, x='total_bill', y='tip', hue='sex')

# also a legend is created by default
```

```python
# style
# number of dinner and lunch

sns.scatterplot(tips, x='total_bill', y='tip', hue='sex', style='time')
```

```python
# size
sns.scatterplot(tips, x='total_bill', y='tip', hue='sex', style='time', size='size')
```

```python
# scatter plot -> figure level function -> square shaped -> recommended by seaborn

sns.relplot(tips, x='total_bill', y='tip', kind='scatter')
```

```python
sns.relplot(tips, x='total_bill', y='tip', kind='scatter', hue='sex')
```

```python
sns.relplot(tips, x='total_bill', y='tip', kind='scatter', hue='sex', style='time')
```

```python
sns.relplot(tips, x='total_bill', y='tip', kind='scatter', hue='sex', style='time', size='size')
```
